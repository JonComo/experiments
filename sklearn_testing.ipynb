```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = np.linspace(0, 10, 100)
data.shape = (len(data), 1)
label = data*2 + np.random.ranf(data.shape)*10.0

model = linear_model.LinearRegression()
model.fit(X=data, y=label)
model.decision_function(data**2)

plt.scatter(data, label)
plt.plot(data, model.predict(data), 'r')
```
