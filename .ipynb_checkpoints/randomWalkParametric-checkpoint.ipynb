```python
import numpy as np
from matplotlib import pyplot as plt
import IPython.display as display
```

```python
t = np.linspace(0, 10 * np.pi, 100)
l = np.sin(t)
```

```python
plt.plot(t, l)
```

```python
waves = 20
w = np.random.random_sample((waves*3)) * .1 - .05 # freq, phase, amp
prevCost = float('+inf')
```

```python
for i in xrange(100000):
    wt = w + np.random.random_sample((waves*3)) *.5 - .25
    h = np.zeros_like(t)
    for j in xrange(waves):
        h += np.sin(wt[j*3] * t + wt[j*3+1]) * wt[j*3+2]
    
    if i%10000==0:
        display.clear_output(wait=True)
        plt.plot(t, l, 'b-')
        plt.plot(t, h, 'r-')
        plt.show()
    
    cost = np.sum((l - h)**2)
    if cost < prevCost:
        prevCost = cost
        w = wt
```
