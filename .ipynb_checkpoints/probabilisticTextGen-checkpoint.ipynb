```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
import wikipedia
s = wikipedia.summary("USA", sentences=20)
print s

# s = "This is my word, and you will take it as law. I write the truth and can only write the truth. Peace."

vocab = set(s)
m = {} # Mapping

for i, c in enumerate(s):
    if i == len(s)-1:
        break
        
    nextC = s[i+1]
    if c not in m:
        m[c] = {}
    
    if nextC in m[c]:
        m[c][nextC] += 1
    else:
        m[c][nextC] = 1

# Convert to probabilities, or use softmax
# for c in m.iterkeys():
#     s = sum(m[c].values())
#     for nextC in m[c].iterkeys():
#         m[c][nextC] = float(m[c][nextC]) / s
```

```python
# Generate a string using these probabilities
newS = "A"
lastC = newS[-1]
temp = 1.0

for __ in xrange(100):
    vals = np.array(m[lastC].values(), dtype=np.float64)
    e = np.exp(vals * temp)
    dist = e/np.sum(e)
    plt.plot(dist)
    nextC = np.random.choice(m[lastC].keys(), p=dist)
#     if nextC in m[lastC]:
#         m[lastC][nextC] = max(0, m[lastC][nextC]-1)
        
    newS += nextC
    lastC = nextC
    if lastC not in m:
        break
    
print newS
```
