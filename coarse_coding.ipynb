```python
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import clear_output
```

```python
def target_fn(x):
    return x * (x > 0)
    # return np.sin(x)
```

```python
T = np.linspace(-5, 5, 100)

f = plt.plot(T, target_fn(T))
```

```python
def basis(x, c, r):
    if c - .5 * r < x < c + .5 * r:
        return 1
    return 0
```

```python
T = np.linspace(-5, 5, 100)
y = [basis(t, 0, 1) for t in T]

f = plt.plot(T, y)
```

```python
def weighted_sum(x):
    s = 0.0
    for i, w in enumerate(W):
        s += w * basis(x, CENTERS[i], WIDTH)
    return s
```

```python
NUM_CENTERS = 100
WIDTH = 1.0
CENTERS = np.linspace(-5, 5, NUM_CENTERS)
W = np.random.random_sample((NUM_CENTERS))



T = np.linspace(-5, 5, 100)
y = [weighted_sum(t) for t in T]

f = plt.plot(T, y)
```

```python
for i in range(1000):
    x = np.random.random_sample() * 10 - 5
    d1 = weighted_sum(x) - target_fn(x)
    d2 = np.array([basis(x, c, WIDTH) for c in CENTERS])
    gW = d1 * d2
    W -= 0.1 * gW
    
    # Plot
    if i % 100 == 0:
        clear_output(wait=True)
        T = np.linspace(-5, 5, 100)
        y = [weighted_sum(t) for t in T]

        f = plt.plot(T, y)
        plt.plot(T, target_fn(T))
        plt.show()
```
